```python
import matplotlib.pyplot as plt
```

```python
def speedup(arr):
  return [arr[0] / e for e in arr]

def eff(arr, cores):
  return [e / c for e, c in zip(speedup(arr), cores)]

def draw(par, eth):
  plt.figure(figsize=(15, 7))
  plt.plot(cores, speedup(par), label='par')
  plt.plot(cores, speedup(eth), label='eth')
  plt.legend()
  plt.show()

  plt.figure(figsize=(15, 7))
  plt.plot(cores, eff(par, cores), label='par')
  plt.plot(cores, eff(eth, cores), label='eth')
  plt.legend()
  plt.show()
  pass

cores = [1, 2, 3, 4]
```

```python
# t1

par = [973.097, 645.827, 510.489, 492.725]
eth = [499.707, 307.549, 244.177, 296.182]

draw(par, eth)
```
